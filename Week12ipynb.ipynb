```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import math
from google.colab.patches import cv2_imshow
from PIL import Image as im
```

```python
cascade_wajah = cv.CascadeClassifier('/content/drive/MyDrive/PCVK(gambar)/pt12/haarcascade_frontalface_alt.xml') 

jokowi = cv.imread('/content/drive/MyDrive/PCVK(gambar)/pt12/jokowi.jpg',0)
roi_wajah = cascade_wajah.detectMultiScale(jokowi)
for(x,y,w,h) in roi_wajah:
  cv.rectangle(jokowi,(x,y),(x+w,y+h),(255,255,255),3)
plt.figure(figsize = (10,10))
plt.imshow(jokowi, cmap='gray')
```
